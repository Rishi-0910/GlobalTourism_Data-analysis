# Data visualization_02

### Importing Libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

### Dataset loading 

```python
Clean_DataSet=pd.read_csv(r"C:\Users\Rishik Reddy\OneDrive\Desktop\Travelnfo\DataSets\ProcessedData\Merged_Tourism_Dataset.csv")
Clean_DataSet.head() 
```

## 1) Global Tourist Arrivals Trend

```python
arrivals = Clean_DataSet.groupby("Year")["Arrivals"].sum()

fig, ax = plt.subplots(figsize=(12,7))

sns.lineplot(x=arrivals.index, y=arrivals.values, marker="o", ax=ax)

ax.set_title("Global Tourist Arrivals Over Years")
ax.set_xlabel("Year")
ax.set_ylabel("Total Arrivals")

ax.grid(True, linestyle="--", alpha=0.6)

plt.show()
```

## Graph summary 
* The line chart represents the trend of global tourist arrivals over the years. It helps identify whether international tourism is increasing, decreasing, or fluctuating.<br>

* An upward trend suggests growth in the tourism industry, while sudden declines may indicate global events such as economic crises or travel restrictions that affect tourism demand.

## 2) Top Countries by Tourism Growth

```python
growth = Clean_DataSet.groupby("Country Name")["TourismGrowth"].mean().sort_values(ascending=False).head(10)
fig, ax = plt.subplots(figsize=(12,7))
growth.plot(kind="bar", ax=ax, color="orange")

ax.set_title("Top 10 Countries by Tourism Growth")
ax.set_xlabel("Country")
ax.set_ylabel("Average Tourism Growth")
plt.xticks(rotation=45)

ax.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

## Graph summary

* This bar chart highlights the countries experiencing the fastest tourism growth. It helps identify emerging tourism destinations where the number of visitors is increasing rapidly.<br>

* These countries may be improving tourism infrastructure, marketing strategies, or accessibility, leading to stronger growth in their tourism sector.<br>

## 3) Top Countries by Tourism Revenue

```python
remove_list = [
"Post-demographic dividend",
"IDA & IBRD total",
"IBRD only",
"Euro area",
"Early-demographic dividend",
"Late-demographic dividend",
"IDA total",
"Small states",
"Fragile and conflict affected situations",
"Pre-demographic dividend",
]
Clean_DataSet=Clean_DataSet[~Clean_DataSet["Country Name"].isin(remove_list)]
```

```python
top_revenue = Clean_DataSet.groupby("Country Name")["Expenditure"].sum().sort_values().tail(10)
fig, ax = plt.subplots(figsize=(12,7))

top_revenue.plot(kind="barh", ax=ax, color="teal")
ax.set_title("Top 10 Countries by Tourism Revenue")
ax.set_xlabel("Total Tourism Expenditure")
ax.set_ylabel("Country")

ax.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

## Graph summary 

* The horizontal bar chart shows the countries generating the highest tourism revenue through visitor spending. It indicates which tourism markets contribute the most to the global tourism economy.<br>

* Countries with higher expenditure levels often offer premium travel experiences, luxury tourism services, or attract high-spending international visitors.

## 4) Expenditure Distribution

```python
top = Clean_DataSet.groupby("Country Name")["Arrivals"].sum().sort_values(ascending=False).head(5)

fig, ax = plt.subplots(figsize=(8,8))
ax.pie(
    top.values,
    labels=top.index,
    autopct="%1.1f%%",
    startangle=90
)

ax.set_title("Tourism Share by Top 5 Countries")
plt.show() 
```

## Graph summary 

* The pie chart illustrates the proportion of global tourist arrivals contributed by the top five countries. It helps understand how tourism demand is distributed among leading destinations.<br>

* Larger slices represent countries attracting a greater share of international tourists.


## 5) **Tourism Market Analysis:** Arrivals, Revenue, and Spending Patterns

```python
sns.pairplot(Clean_DataSet[["Arrivals","Expenditure","TourismGrowth","AvgSpendingPerTourist"]])
plt.Figure(figsize=(12,7))
plt.show()
```

## Graph summary 

* The pairplot visualizes relationships between multiple tourism variables. It helps identify correlations between tourist arrivals, tourism expenditure, tourism growth, and average spending per tourist.<br>

* This type of visualization is useful for detecting patterns, clusters, and potential relationships between tourism indicators.
