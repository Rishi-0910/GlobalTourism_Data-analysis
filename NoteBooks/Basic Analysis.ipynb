## 1️ Arrivals Data (ArrivalsData.csv)

### Indicator
- Name: International Tourism – Number of Arrivals
- Code: ST.INT.ARVL

### Key Points
- Contains number of international tourist arrivals.
- Covers multiple countries.
- Time period: 1960 – 2025.
- Useful to analyze tourism growth trends.
- Helps identify impact of global events (e.g., COVID-19 decline).
- Can be used for forecasting tourism demand.
* **Link:**[World Bank - Tourism Arrival ] *https://data.worldbank.org/indicator/ST.INT.ARVL*
---

## 2️ Expenditure Data (ExpenditureData.csv)

### Indicator
- Name: International Tourism – Expenditure

### Key Points
- Shows total spending by international tourists.
- Helps measure tourism-related economic activity.
- Enables country-wise expenditure comparison.
- Useful for revenue trend analysis.
- Supports economic impact studies.
- **Link:**[World Bank – Tourism Expenditure ] *https://data.worldbank.org/indicator/ST.INT.XPND.CD*

---

## 3️ Receipts Data (ReceiptsData.csv)

### Indicator
- Name: International Tourism – Receipts

### Key Points
- Represents income earned from foreign tourists.
- Measures tourism contribution to national economy.
- Helps analyze foreign exchange earnings.
- Useful for GDP contribution studies.
- Supports financial performance comparison across countries.
- **Link:**[World Bank – Tourism Receipts ] *https://data.worldbank.org/indicator/ST.INT.RCPT.XP.ZS*

## Remove Unnecessary Columns

```python
cols_to_drop = ['Indicator Name', 'Indicator Code']
arrivals = arrivals.drop(columns=cols_to_drop)
expenditure = expenditure.drop(columns=cols_to_drop)
receipts = receipts.drop(columns=cols_to_drop)
```

```python
arrivals.columns
```

## Keep Relevant Years

```python
years = [str(year) for year in range(2005, 2024)]

arrivals = arrivals[['Country Name','Country Code'] + years]
expenditure = expenditure[['Country Name','Country Code'] + years]
receipts = receipts[['Country Name','Country Code'] + years]
```

```python
arrivals.columns
```

## Remove Aggregated Regions

```python
remove_words = "World|income|OECD|Asia|Europe|Africa|America"

arrivals = arrivals[~arrivals['Country Name'].str.contains(remove_words)]
expenditure = expenditure[~expenditure['Country Name'].str.contains(remove_words)]
receipts = receipts[~receipts['Country Name'].str.contains(remove_words)]
```

```python
arrivals.info()
```

## Convert from Wide → Long Format

```python
arrivals_long = arrivals.melt(
    id_vars=['Country Name','Country Code'],
    var_name='Year',
    value_name='Arrivals'
)

expenditure_long = expenditure.melt(
    id_vars=['Country Name','Country Code'],
    var_name='Year',
    value_name='Expenditure'
)
```

```python
arrivals_long.head()
```

## Merge Datasets

```python
merged = arrivals_long.merge(
    expenditure_long,
    on=['Country Name','Country Code','Year'],
    how='inner'
)
```

```python
merged.head()
```
