# Data Visualization_03

### Importing libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

### Loading dataset

```python
Clean_DataSet=pd.read_csv(r"C:\Users\Rishik Reddy\OneDrive\Desktop\Travelnfo\DataSets\ProcessedData\Merged_Tourism_Dataset.csv")
Clean_DataSet.head()
```

## 1) Correlation Analysis of Global Tourism Indicators

```python
corr = Clean_DataSet[["Arrivals","Expenditure","TourismGrowth","AvgSpendingPerTourist"]].corr()

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Between Tourism Variables")
plt.xticks(rotation=0)
plt.show()
```

| Value  | Meaning                       |
| ------ | ----------------------------- |
| **+1** | Perfect positive relationship |
| **0**  | No relationship               |
| **-1** | Perfect negative relationship |


## Graph summary 

* This heatmap shows the relationships between key tourism variables such as tourist arrivals, expenditure, tourism growth, and average spending per tourist.<br>

* The visualization indicates that tourist arrivals and tourism expenditure have a strong positive correlation, meaning countries with higher visitor numbers generally generate more tourism revenue.<br>

* Other variables show weaker relationships, suggesting that growth rates and average spending may vary independently across destinations.

## 2) Yearly Distribution of International Tourist Arrivals

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="Year", y="Arrivals", data=Clean_DataSet)
plt.title("Distribution of Tourist Arrivals by Year")
plt.xticks(rotation=90)
plt.show()
```

## Graph summary 

* This box plot illustrates how tourist arrivals vary across countries for each year. Most countries receive relatively low numbers of tourists, while a few major destinations attract extremely high arrivals, appearing as outliers.<br>

* The distribution also suggests a gradual increase in maximum arrivals over time before a noticeable decline in 2020, reflecting the global impact on tourism.

## 3) Distribution of International Tourist Arrivals Across Countries 

```python
plt.figure(figsize=(10,6))
sns.histplot(Clean_DataSet["Arrivals"], bins=40, kde=True)
plt.title("Distribution of Tourist Arrivals")
plt.show()
```

## Graph summary 

* The histogram reveals that tourist arrivals are heavily concentrated at lower values, meaning most countries receive modest numbers of visitors.<br>

* A long right tail indicates that a small number of countries attract very large tourist volumes. This pattern highlights the unequal global distribution of tourism activity.<br>

## 4) Tourist Arrivals vs Tourism Expenditure with Average Spending Impact

```python
plt.figure(figsize=(10,6))

plt.scatter(
    Clean_DataSet["Arrivals"],
    Clean_DataSet["Expenditure"],
    s=Clean_DataSet["AvgSpendingPerTourist"]*0.1,
    alpha=0.6
)

plt.xlabel("Arrivals")
plt.ylabel("Expenditure")
plt.title("Tourism Arrivals vs Expenditure (Bubble = Avg Spending)")
plt.show()
```

## Graph summary 

* This bubble chart demonstrates the relationship between tourist arrivals and tourism expenditure. The upward trend indicates a strong positive relationship, where countries with higher tourist arrivals tend to generate greater tourism revenue.<br>

* The varying bubble sizes reveal differences in average spending per tourist, showing that some destinations earn more per visitor than others.

## 5) Tourism Arrival Trends for Top Performing Countries Over Time

```python
remove_list = [
"Post-demographic dividend",
"IDA & IBRD total",
"IBRD only",
"Euro area",
"Early-demographic dividend",
"Late-demographic dividend",
"IDA total",
"Small states",
"Fragile and conflict affected situations",
"Pre-demographic dividend",
]
Clean_DataSet=Clean_DataSet[~Clean_DataSet["Country Name"].isin(remove_list)]
```

```python
top_countries = Clean_DataSet.groupby("Country Name")["Arrivals"].sum().nlargest(5).index

filtered = Clean_DataSet[Clean_DataSet["Country Name"].isin(top_countries)]

plt.figure(figsize=(12,6))
sns.lineplot(data=filtered, x="Year", y="Arrivals", hue="Country Name")
plt.title("Tourism Arrivals Trend for Top 5 Countries")
plt.show()
```

## Graph summary 

* This line chart tracks the tourism arrival trends for the top countries over several years. The visualization shows consistent growth in tourist arrivals for leading destinations until recent years, highlighting their dominance in global tourism.<br>

* Fluctuations in later years suggest external influences that impacted international travel patterns.
