### cheacking NULL values 

```python
MergedDataset.isnull().sum()
```

```python
MergedDataset['Growth_%'] = MergedDataset.groupby('Country Name')['Arrivals'].pct_change()*100
MergedDataset.head()
```

```python
MergedDataset.rename(columns={'Growth_%': 'TourismGrowth'}, inplace=True)
MergedDataset.head(500)

# NOTE : The first year for each country will show NAN beacause there is no previous year to compare to calculate the growth percentage
```

```python

MergedDataset['TourismGrowth'] = (
    MergedDataset['TourismGrowth']
    .astype(str)
    .str.replace('%', '', regex=False)
)
MergedDataset['TourismGrowth'] = pd.to_numeric(MergedDataset['TourismGrowth'], errors='coerce')
MergedDataset['TourismGrowth'].dtypes
```

```python
avg_growth = MergedDataset['TourismGrowth'].mean()
print("Average Tourism Growth:", avg_growth)
```

### Spending Per Tourist

```python
MergedDataset['AvgSpendingPerTourist'] = MergedDataset['Expenditure']/MergedDataset['Arrivals']
MergedDataset.head()
```

###  Top 10 Countries by Tourism Revenue

```python
remove_words = "income|dividend|IBRD|IDA|Euro|OECD|area|total"

merged = MergedDataset[~MergedDataset['Country Name'].str.contains(remove_words, case=False, na=False)]
merged.groupby('Country Name')['Expenditure'] \
      .sum() \
      .sort_values(ascending=False) \
      .head(10)
```

### COVID Impact (2019–2021)

```python
MergedDataset[MergedDataset['Year'].isin([2019,2020,2021])]
```

```python
data_2019= MergedDataset[MergedDataset['Year'] == 2019]['Arrivals'].mean()
data_2020= MergedDataset[MergedDataset['Year'] == 2020]['Arrivals'].mean()
data_2021= MergedDataset[MergedDataset['Year'] == 2021]['Arrivals'].mean()
print("Average Arrivals in 2019:", data_2019.round(2))
print("Average Arrivals in 2020:", data_2020.round(2))  
print("Average Arrivals in 2021:", data_2021)
```

```python
drop_percent = ((data_2020 - data_2019) / data_2019) * 100
print("Total arrivals from 2019 to 2020:",drop_percent.round(2), "%")
```
