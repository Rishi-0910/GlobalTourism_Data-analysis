# Data visualization_01

### Importing Libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

### Dataset loading 

```python
Clean_DataSet=pd.read_csv(r"C:\Users\Rishik Reddy\OneDrive\Desktop\Travelnfo\DataSets\ProcessedData\Merged_Tourism_Dataset.csv")
Clean_DataSet.head()
```

## 1) Tourist Arrivals Over Time

```python
figure, ax = plt.subplots(figsize=(10, 6))

ax.plot(Clean_DataSet["Year"], Clean_DataSet["Arrivals"])

ax.set_title("Tourist Arrivals Over Years")
ax.set_xlabel("Year")
ax.set_ylabel("Arrivals")

ax.grid(True)   

plt.show()
```

### Graph summary 
* This visualization shows the trend of tourist arrivals across different years. The line graph helps identify whether tourism is increasing, decreasing, or fluctuating over time.<br>

* A rising trend indicates growth in the tourism industry, while drops may indicate global events, economic issues, or travel restrictions affecting tourism demand.


## 2) Top 10 Tourist Destination Countries

```python
remove_list = [
"Post-demographic dividend",
"IDA & IBRD total",
"IBRD only",
"Euro area",
"Early-demographic dividend",
"Late-demographic dividend",
"IDA total",
"Small states"
]
Clean_DataSet=Clean_DataSet[~Clean_DataSet["Country Name"].isin(remove_list)]
```

```python
figure, ax = plt.subplots(figsize=(10, 6))


Clean_DataSet= Clean_DataSet.groupby("Country Name")["Arrivals"].sum().sort_values(ascending=False).head(10)

Clean_DataSet.plot(kind="bar", ax=ax)
ax.set_title("Top 10 Tourist Destinations")
ax.set_ylabel("Total Arrivals")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
plt.show()
```

### Graph summary 
* This bar chart represents the top 10 tourist destination countries based on total tourist arrivals. It highlights which countries dominate global tourism and attract the highest number of visitors.<br>

* Countries with higher bars are major tourism hubs due to factors such as cultural attractions, infrastructure, economic development, and international accessibility.

## 3) Tourism Spending vs Tourist Arrivals

```python

figure, ax = plt.subplots(figsize=(12,7))

sns.scatterplot(
    data=Clean_DataSet,
    x="Arrivals",
    y="Expenditure",
    ax=ax,
    alpha=0.6
)

ax.set_title("Tourism Spending vs Tourist Arrivals")
ax.set_xlabel("Tourist Arrivals")
ax.set_ylabel("Tourism Expenditure")

ax.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

### Graph summary 
* This scatter plot visualizes the relationship between the number of tourist arrivals and tourism expenditure. Each point represents a country or data observation<br>

* If the points trend upward, it indicates that higher tourist arrivals generally lead to higher tourism spending.<br>

* The distribution of points helps identify correlations, clusters, or outliers in tourism spending behavior.

## 4) Tourism Growth Distribution

```python
growth = Clean_DataSet.groupby("Year")["TourismGrowth"].mean()

fig, ax = plt.subplots(figsize=(12,7))

sns.lineplot(x=growth.index, y=growth.values, marker="o", ax=ax)
ax.set_title("Average Tourism Growth Over Years")
ax.set_xlabel("Year")
ax.set_ylabel("Tourism Growth")

ax.grid(True, linestyle="--", alpha=0.6)

plt.show()
```

### Graph summary 
* This visualization shows the average tourism growth rate over time. It helps identify years with strong tourism expansion or economic decline.<br>

* Peaks represent years with high tourism growth, while dips indicate periods where tourism slowed down due to economic, political, or global events.

## 5)Average Spending Per Tourist by Country

```python
avg_spending = Clean_DataSet.groupby("Country Name")["AvgSpendingPerTourist"].mean().sort_values(ascending=False).head(10)

fig, ax = plt.subplots(figsize=(12,7))
ax.fill_between(avg_spending.index, avg_spending.values, color="skyblue", alpha=0.6)

ax.plot(avg_spending.index, avg_spending.values, color="blue", marker="o")

ax.set_title("Top 10 Countries by Average Spending Per Tourist")
ax.set_xlabel("Country")
ax.set_ylabel("Average Spending Per Tourist")
plt.xticks(rotation=45)
ax.grid(True, linestyle="--", alpha=0.6)

plt.show()
```

### Graph summary 
* This visualization displays the countries where tourists spend the most money on average. It helps identify high-value tourism markets where each visitor contributes more to the local economy.<br>

* Countries with higher spending per tourist typically offer premium tourism experiences, luxury services, or high-cost travel destinations.
