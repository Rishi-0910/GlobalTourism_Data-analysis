# Data Visualization_04

### Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

### Loading Dataset

```python
Clean_DataSet=pd.read_csv(r"C:\Users\Rishik Reddy\OneDrive\Desktop\Travelnfo\DataSets\ProcessedData\Merged_Tourism_Dataset.csv")
Clean_DataSet.head()
```

## 1) Global Tourism Trend Over Time

```python
yearly = Clean_DataSet.groupby("Year")["Arrivals"].sum()

plt.figure(figsize=(10,6))
plt.plot(yearly.index, yearly.values, marker='o')
plt.title("Global Tourist Arrivals Trend Over Time")
plt.xlabel("Year")
plt.ylabel("Total Arrivals")
plt.show()
```

## Graph summary

- **Global tourist arrivals show a clear long-term trend**, indicating how international travel demand has evolved over the years.<br> 

- **Certain years display noticeable growth**, suggesting improvements in global connectivity and tourism infrastructure.<br>  
- **Minor fluctuations indicate economic or global influences** that temporarily affect tourism movement.<br>  
- Overall, the graph highlights the **general expansion of global tourism activity over time**.

## 2) Top 10 Countries by Tourist Arrivals

```python
remove_list = [
"Post-demographic dividend",
"IDA & IBRD total",
"IBRD only",
"Euro area",
"Early-demographic dividend",
"Late-demographic dividend",
"IDA total",
"Small states",
"Fragile and conflict affected situations",
"Pre-demographic dividend",
]
Clean_DataSet=Clean_DataSet[~Clean_DataSet["Country Name"].isin(remove_list)]
```

```python
top10 = Clean_DataSet.groupby("Country Name")["Arrivals"].sum().nlargest(10)

plt.figure(figsize=(10,6))
top10.sort_values().plot(kind="barh")
plt.title("Top 10 Countries by Tourist Arrivals")
plt.xlabel("Total Arrivals")
plt.show()
```

## Graph summary

- The chart identifies the **top countries attracting the highest number of international tourists**.<br>

- A **small group of countries dominates global tourism**, showing strong travel appeal and infrastructure.<br>

- These destinations likely benefit from **popular attractions, accessibility, and tourism marketing**.<br>

- The visualization highlights the **global concentration of tourist traffic among major destinations**.<br>

## 3) Country Tourism Pattern

```python
plt.figure(figsize=(10,8))
pivot = Clean_DataSet.pivot_table(
    values="Arrivals",
    index="Country Name",
    columns="Year",
    aggfunc="sum"
)
pivot.fillna(0, inplace=True)

sns.heatmap(pivot.head(10), cmap="YlGnBu")
```

## Graph summary

- The heatmap visualizes **tourist arrivals across countries and years simultaneously**.<br>

- Darker color intensity represents **higher tourism activity**, helping identify strong tourism performers.<br>

- Some countries maintain **consistently high tourism levels**, while others show gradual growth over time.<br> 

- This visualization makes it easy to **compare tourism performance across multiple countries and time periods**.<br>



## 4) Tourism Market Structure

```python
plt.figure(figsize=(10,8))
share = Clean_DataSet.groupby("Country Name")["Arrivals"].sum().nlargest(5)
share.plot(kind="pie", autopct="%1.1f%%")
```

## Graph summary 
- The pie chart displays the **share of global tourist arrivals among the top tourism countries**.<br>

- A few destinations hold **a large proportion of the total tourism market**.<br>

- This indicates that **global tourism demand is heavily concentrated in popular destinations**.<br>

- The visualization highlights the **market dominance of leading tourism economies**.<br>
